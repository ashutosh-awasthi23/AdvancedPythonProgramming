```python
import pandas as pd
```

```python
students_df=pd.read_csv("students.csv")
students_df
```

```python
extra_scores_df=pd.read_csv("scores.csv")
extra_scores_df
```

```python

#cleaning
students_df.dropna(subset=["Age"],inplace=True)
students_df
```

```python
students_df["Score"].fillna(students_df["Score"].mean(),inplace=True)
students_df
```

```python
#Prepare data
students_df[(students_df["Score"]>70) & (students_df["Age"]>20)]
```

```python
students_df["Final_Score"]=students_df["Score"].astype(int)+10
```

```python
students_df.drop(columns=["Gender"],inplace=True)
```

```python
students_df
```

```python
students_df.sort_values(by=["Final_Score"],ascending=[False],inplace=True)
```

```python
students_df
```

```python
df=pd.merge(students_df,extra_scores_df,on="ID")
df
```

```python
df["Total_Score"]=df["Final_Score"]+df["Additional_Score"]
df
```
